```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup
prefix = "https://www.histdata.com"
currencys = {}
final = {}
data = urlopen("https://www.histdata.com/download-free-forex-data/?/ascii/1-minute-bar-quotes")
soup = BeautifulSoup(data, 'html.parser')
table = soup.find("table")
for link in table.find_all("a"):
    currencys[link.attrs["title"][61:67]] = link.attrs["href"]
    final[link.attrs["title"][61:67]] = []
for currency in currencys:
    data = urlopen(prefix + currencys[currency])
    soup = BeautifulSoup(data, 'html.parser')
    for link in soup.find_all("p")[3].find_all("a"):
        final[currency].append({link.attrs["title"][87:] : link.attrs["href"]})
    
```
